```python
#
# Vectorize teams
#
ret = None 

for cat in categorical_cols:
  labs = ['A', 'B','C', 'D', 'E', 'F', 'G', 'H']
  labs.remove(cat)
  cur = pd.get_dummies(tms, columns=[cat], prefix="Hero_")
  cur = cur.drop(labels=labs,  axis=1)

  if ret is None:
    ret = cur
  else:
    ret = ret.add(cur, fill_value=0)
```
